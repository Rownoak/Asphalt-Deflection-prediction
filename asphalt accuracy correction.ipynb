```python
import pandas as pd
import numpy as np
import warnings as wr #Ignores the warnings
wr.filterwarnings('ignore')
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from pandas.core.dtypes.common import is_numeric_dtype #"from pandas.core.dtypes.common import is_numeric_dtype" imports the is_numeric_dtype function from the pandas.core.dtypes.common module. This function is used to check if a given pandas data type is numeric or not.
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

import seaborn as sns
```

```python
 df= pd.read_csv("dataset.csv")
```

```python
df.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
train=df.drop(['Deflection (mm)'],axis=1)
test=df['Deflection (mm)']
```

```python
x_train, x_test, y_train, y_test = train_test_split(train, test,test_size=0.3, random_state=1)
```

```python
regr=LinearRegression()
```

```python
regr.fit(x_train,y_train)
```

```python
pred=regr.predict(x_test)
```

```python
pred
```

```python
regr.score(x_test,y_test)
```

```python
# Test data
test_data = np.array([[0.8,1,2]])

# Predict the values for the test data
predicted_values = regr.predict(test_data)

# Print the predicted values
print(f"Predicted value for test data 0.8: {predicted_values[0]}")
```

```python
# Scatter plot of the data points
plt.scatter(x_test['Temperature (C)'].sort_values(), y_test, color='blue', label='Actual')

# Plot the linear regression line
plt.plot(x_test['Temperature (C)'].sort_values(), pred, color='red', label='Linear Regression')

# Set labels and title
plt.xlabel('Temperature (C)')
plt.ylabel('Deflection (mm)')
plt.title('Scatter Plot for Linear Regression')

# Add legend
plt.legend()

# Show the plot
plt.show()

# Calculate the accuracy (R-squared score) on the test data
accuracy = regr.score(x_test, y_test)
print(f"Accuracy (R-squared score): {accuracy}")
```

```python
# Scatter plot of the data points
plt.scatter(x_test['Force (mN)'].sort_values(), y_test, color='blue', label='Actual')

# Plot the linear regression line
plt.plot(x_test["Force (mN)"].sort_values(), pred, color='red', label='Linear Regression')

# Set labels and title
plt.xlabel('Force (mN)')
plt.ylabel('Deflection (mm)')
plt.title('Scatter Plot for Linear Regression')

# Add legend
plt.legend()

# Show the plot
plt.show()

# Calculate the accuracy (R-squared score) on the test data
accuracy = regr.score(x_test, y_test)
print(f"Accuracy (R-squared score): {accuracy}")
```
